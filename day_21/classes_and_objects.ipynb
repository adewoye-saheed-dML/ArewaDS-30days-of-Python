```python
# creating a statistics class which can calculate the measure of central tendency 
# of a sample and measure of variability.


class statistics:
    
    def __init__(self,age):
        self.age = age
        
    def mean(self):
        for i in self.age:
            mean = sum(self.age) / len(self.age)
        return f'{mean:.0f}'
        
    def count(self):
        count= len(list(i for i in self.age if isinstance(i, int)))
        return count

    def median(self):
        for i in self.age:
            self.age.sort()
            mid = len(self.age) // 2
            median = (self.age[mid] + self.age[~mid]) /2
        return median

    def mode(self):
        for i in self.age:
            mode = max(self.age, key=self.age.count)
        return mode
        
    def range(self):
        self.age.sort()
        for i in self.age:
            range = self.age[-1] - self.age[0]
        return range

    def min(self):
        for i in self.age:
            self.age.sort()
            mini = self.age[0]
        return mini

    def max(self):
        for i in self.age:
            self.age.sort()
            maxi = self.age[-1]
        return maxi

    def sum(self):
        total = 0
        for i in self.age:
            total+=i
        return total
        
    def var(self):
        sum_squared_diff = 0             
        for i in self.age:
            mean = sum(self.age) / len(self.age)
            sum_square = (i - mean) ** 2
            sum_squared_diff += sum_square
            var = sum_squared_diff / len(self.age)
        return f'{var:.1f}'
        
    def std(self):
        sum_squared_diff = 0
        for i in self.age:
            mean = sum(self.age) / len(self.age)
            sum_square = (i - mean) ** 2
            sum_squared_diff += sum_square
            var = sum_squared_diff / len(self.age)
            std = var ** 0.5
        return f'{std:.1f}'

data = statistics(age = [31, 26, 34, 37, 27, 26, 32, 32, 26, 27, 27, 24, 32, 33, 27, 25, 26, 38, 37, 31, 34, 24, 33, 29, 26])

print('Count:', data.count()) 
print('Sum: ', data.sum()) 
print('Min: ', data.min()) 
print('Max: ', data.max()) 
print('Range: ', data.range())
print('Mean: ', data.mean())
print('Median: ', data.median()) 
print('Mode: ', data.mode()) 
print('Standard Deviation: ', data.std())
print('Variance: ', data.var()) 
        
```

```python
# creating a person account class

class PersonAccount:
    def __init__(self, firstname, lastname):
        self.firstname = firstname
        self.lastname = lastname
        self.incomes = {}
        self.expenses = {}

    def add_income(self, description, amount):
        if description in self.incomes:
            self.incomes[description] += amount
        else:
            self.incomes[description] = amount

    def add_expense(self, description, amount):
        if description in self.expenses:
            self.expenses[description] += amount
        else:
            self.expenses[description] = amount

    def total_income(self):
        return sum(self.incomes.values())

    def total_expense(self):
        return sum(self.expenses.values())

    def account_balance(self):
        return self.total_income() - self.total_expense()

    def account_info(self):
        print(f"Account Information for {self.firstname} {self.lastname}:")
        print(f"Total Income: ${self.total_income()}")
        print(f"Total Expense: ${self.total_expense()}")
        print(f"Account Balance: ${self.account_balance()}")
        print("\nIncome Details:")
        for description, amount in self.incomes.items():
            print(f"{description}: ${amount}")
        print("\nExpense Details:")
        for description, amount in self.expenses.items():
            print(f"{description}: ${amount}")


# Example Usage:
person1 = PersonAccount("Adewoye", "Saheed")

person1.add_income("Salary", 5000)
person1.add_income("Bonus", 1000)

person1.add_expense("Rent", 1500)
person1.add_expense("Groceries", 300)
person1.add_expense("Utilities", 200)

person1.account_info()
```
